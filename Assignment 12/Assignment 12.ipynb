Use <em>.groupby('columnname').mean()</em> to calculate the means. Is there a difference between categories? Then use seaborn barplots to check if there is a statistically significant difference. 

```python
import pandas as pd
import seaborn as sns
```

```python
stockx = pd.read_excel('StockX_Dataset.xlsx')
```

```python
stockx.head()
```

```python
stockx.groupby(by="Brand").mean()
```

```python
sns.barplot(y="Sale Price", x="Brand", data=stockx)
```

Er een een statistically significant difference bij Sale Price tussen de Brands Yeezy en Off-White. Dit is ook zo in de werkelijkheid. Off-White sneakers zijn vaak vele malen duurder dan Yeezy's dus ik had ook verwacht dat deze grafiek er zo uit zal zien.

```python
sns.barplot(y="Retail Price", x="Brand", data=stockx)
```

Deze grafiek is het tegenovergestelde van de Sale Price grafiek. Ook hier is een statistically significant difference. De retail price is bij Yeezy's vaak hoger dan die van Off-White sneakers. Dit is te zien in de onderstaande grafieken:

```python
filter = stockx["Brand"] == "Yeezy"
stockx[filter]['Retail Price'].value_counts().sort_index(ascending=True).plot(kind='bar')
```

```python
filter = stockx["Brand"] == "Off-White"
stockx[filter]['Retail Price'].value_counts().sort_index(ascending=True).plot(kind='bar')
```

```python
sns.barplot(y="Shoe Size", x="Brand", data=stockx)
```

Bij de Shoe Sizes zit er een klein verschil en de zwarte lijnen komen niet overeen. Er is dus sprake van statistically significant difference. Dit vond ik wel interessant om te zien dat er schijnbaar vaker grotere maten Off-White sneakers worden gekocht dan Yeezy's.
