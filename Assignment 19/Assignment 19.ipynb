```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins").dropna()
penguins.head()
```

```python
sns.pairplot(penguins)
```

Ik zie dat er 2 clusters zijn. Dit is omdat in de visualisaties echt 2 groepen ontstaan. Bij sommige lijkt het ook alsof er nog een derde cluster erbij komt maar deze is minder groot

```python
features = ['flipper_length_mm', 'body_mass_g']
km = KMeans(n_clusters=2, random_state=43).fit(penguins[features])
```

```python
penguins['cluster'] = km.predict(penguins[features])
```

```python
penguins.head()
```

```python
penguins.cluster.value_counts()
```

```python
sns.pairplot(penguins, hue="cluster")
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
metrics.silhouette_score(penguins[features], km.labels_, metric='euclidean')
```

```python
contingency_table = penguins.groupby(['species','cluster']).size().unstack('cluster', fill_value=0)
contingency_table
```

Er zijn species die in dezelfde cluster vallen. Dit betekent dat er penguins zijn van verschillende species die gelijke waarden hebben.
